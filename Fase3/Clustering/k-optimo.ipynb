```python
import pandas as pd
url ="https://raw.githubusercontent.com/albercol/TFG/main/Fase2/Fase%202.3/DatosUnificados/Cont_Meteo_Traf.csv"
df = pd.read_csv(url,parse_dates=['FECHAH'])

#Se realiza el desplazamiento horario.
df['ES10'] = df['ES10'].shift(1)
df['ES53'] = df['ES53'].shift(1)
df = df.dropna()
df["hora"] = df['FECHAH'].dt.hour

cols = [ 'CO', 'NO2', 'PM2.5', 'O3',
       'VELOCIDAD VIENTO', 'DIR. VIENTO', 'TEMPERATURA', 'HUMEDAD RELATIVA',
       'PRESION BAROMETRICA', 'PRECIPITACIONES', 'ES10', 'ES53']
df2 = df[cols]
df2
```

```python
df
```

Escalamos

#### Normalizacion de los datos.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
datos = scaler.fit_transform(df2)
df3 = pd.DataFrame(columns = df2.columns, data=datos)
df3
```

#### Calculamos el numero de clusters optimo para la agrupaciÃ³n
Uso de kmenas (criterio del codo):

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2).fit(df3)

Sum_of_squared_distances = []
K = range(2,15)
for k in K:
    km = KMeans(n_clusters=k).fit(df3)
    
    Sum_of_squared_distances.append(km.inertia_)
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

plt.plot(range(2,len(Sum_of_squared_distances)+2),Sum_of_squared_distances)
```

Uso de criterio de la silueta:

```python
from sklearn.metrics import silhouette_score

sil = []
kmax = 10
x = df3
# dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in range(2, kmax+1):
  kmeans = KMeans(n_clusters = k).fit(x)
  labels = kmeans.labels_
  #sil.append(silhouette_score(x, labels, metric = 'euclidean'))
  silhouette_value = silhouette_score(x, labels, metric = 'euclidean')
  sil.append(silhouette_value)
  print("Para n_clusters =", k, "el promedio es :", silhouette_value )
```

```python
#AÃ±adimos la columna C (cluster) para identificar a cual pertenece.
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

def get_MAE(df,coly):
    colsX = [  'VELOCIDAD VIENTO', 'DIR. VIENTO', 'TEMPERATURA', 'HUMEDAD RELATIVA',
           'PRESION BAROMETRICA', 'PRECIPITACIONES', 'ES10','ES53']
    X = df[colsX]
    y = df[coly]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
    metodo = RandomForestRegressor()
    modelo = metodo.fit(X_train,y_train)
    y_pred = modelo.predict(X_test)
    X_test["real"] = y_test
    X_test["pred"] = y_pred

    return mean_absolute_error(y_pred,y_test),X_test


for k in range(2,10) :
    km = KMeans(n_clusters=k).fit(df3)
    df["C"] = km.labels_
    print("-------------------------------------------")
    for i in range(0, k) :
        df4 = df[df["C"]==i]
        mae,df_test = get_MAE(df4,"NO2")
        print("Para", k , "Clusters, El mae es de", mae)
```
