```python
import pandas as pd
fichero = "https://raw.githubusercontent.com/albercol/TFG/main/Fase%201.3/Trafico/ArchivoTrafico.csv"
df =  pd.read_csv(fichero,encoding="UTF-8")

# posición de la última columna que queremos
pos_ultima = list(df.columns).index('source_file')
# nos quedamos solo con las columnas que queremos
df = df[df.columns[0:pos_ultima+1]]


# separamos en dos dataframes
total = len(df)
df["sentido"] = df.FSEN.str.get(0)
df["horas"] = df.FSEN.str.get(1)

# OJO vamos a hacer JOIN por una clave, debemos borrar los que sean NULL en esa clave
df.dropna(subset=['FDIA','FEST','sentido'], how='any', inplace=True)
print("Borrados ",total-len(df))

# Por legibilidad separamos en dos ficheros, uno que tiene de 1 a 12 y el otro que tiene de 12 a 24
df1a12 = df[df.horas=="-"]
df13a24 = df[df.horas=="="]
# join
merged_df = pd.merge(df1a12,df13a24,how='outer',on=['FDIA','FEST','sentido'])
                     
# finalmente recolocamos las columnas y las renombramos
resultado = merged_df[['FDIA', 'FEST', 'sentido',  'HOR1_x', 'HOR2_x', 'HOR3_x', 'HOR4_x',
       'HOR5_x', 'HOR6_x', 'HOR7_x', 'HOR8_x', 'HOR9_x', 'HOR10_x', 'HOR11_x',
       'HOR12_x',  'HOR1_y',
       'HOR2_y', 'HOR3_y', 'HOR4_y', 'HOR5_y', 'HOR6_y', 'HOR7_y', 'HOR8_y',
       'HOR9_y', 'HOR10_y', 'HOR11_y', 'HOR12_y', 'source_file_x' ]]
resultado.columns = ['FDIA', 'FEST', 'sentido',  'HOR1', 'HOR2', 'HOR3', 'HOR4',
       'HOR5', 'HOR6', 'HOR7', 'HOR8', 'HOR9', 'HOR10', 'HOR11',
       'HOR12', 'HOR13', 'HOR14', 'HOR15', 'HOR16', 'HOR17', 'HOR18', 'HOR19', 'HOR20',
       'HOR21', 'HOR22', 'HOR23', 'HOR24', 'source_file' ]
resultado.sort_values(by=['FDIA','FEST','sentido'], inplace=True)                         

resultado
```

```python
#Observar que la columna HOR13 y la columna HOR1 es de tipo 'objet' y no 'float64'
print(resultado.dtypes)
#Convertimos los valores de las columnas HOR1 y HOR13 a valores de tipo float.
#(si no se realiza esta conversion, la suma con el groupby realizara una 'exclusión automática de columnas molestas')
resultado["HOR13"] = pd.to_numeric(resultado["HOR13"], errors='coerce')
resultado["HOR1"] = pd.to_numeric(resultado["HOR1"], errors='coerce')

#Realizamos la suma agrupando por FDIA y FEST.
resultado = resultado.groupby(by=["FDIA","FEST"], as_index=False)[['HOR1','HOR2','HOR3','HOR4','HOR5', 
                                                       'HOR6','HOR7','HOR8','HOR9','HOR10','HOR11','HOR12','HOR13','HOR14',
                                                    'HOR15','HOR16','HOR17','HOR18','HOR19','HOR20','HOR21', 
                                                      'HOR22','HOR23','HOR24']].sum()
#Guardamos el resultado en "Traficode12a24.csv"
resultado.to_csv(path+"/Trafico/"+"Traficode12a24.csv", header=True, index=False, encoding="utf-8")
resultado
```
